```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv('Untitled spreadsheet - Amazon Sales data.csv', encoding='unicode_escape')
```

```python
df['Order Date Month'] = pd.to_datetime(df['Order Date Month'])
df['Order Date Year'] = pd.to_datetime(df['Order Date Year'])
df['Ship Date Month'] = pd.to_datetime(df['Ship Date Month'])
df['Ship Date Year'] = pd.to_datetime(df['Ship Date Year'])
```

```python
df.info()
```

```python
Exploratory Data Analysis
```

```python
Region
```

```python
sales_itm = df.groupby(['Region'], as_index=False)['Units Sold'].sum().sort_values(by='Units Sold', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(x = 'Region',y= 'Units Sold' ,data = sales_itm)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_itm = df.groupby(['Region'], as_index=False)['Total Revenue'].sum().sort_values(by='Total Revenue', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(x = 'Region',y= 'Total Revenue' ,data = sales_itm)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_itm = df.groupby(['Region'], as_index=False)['Total Profit'].sum().sort_values(by='Total Profit', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.barplot(x = 'Region',y= 'Total Profit' ,data = sales_itm)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
 From above graphs we can see that most buyers are from Sub-Saharan Africa, Europe, Australia region but purchasing power of Asia is more than Australia.
```
